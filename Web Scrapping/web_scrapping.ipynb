```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
url = "https://www.flipkart.com/search?q=mobiles&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=off&as=off"
r = requests.get(url)
print(r)

soup = BeautifulSoup(r.text, "lxml")
# print(soup)

#boxes = soup.find("div",class_ ="_1YokD2 _3Mn1Gg")
#print(len(boxes))

names = soup.find_all("div",class_ = "_4rR01T")
#for i in names:
 #   print(i.text)

prices = soup.find_all("div",class_ = "_30jeq3 _1_WHN1")
#for j in prices:
 #   print(j.text)

df = pd.DataFrame(columns =['Name','price'])

for name,price in zip(names,prices):
    name_text = name.text.strip()
    price_text = price.text.strip()
    df = df.append({'Name' :name_text,'price' :price_text},ignore_index = True)
df.to_csv('Scraped_data.csv',index = False)
df

```
